```python
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
os.chdir('/users/yoga7i/downloads')
df = pd.read_csv('SYB67_123_202411_Total Imports Exports and Balance of Trade.csv', skiprows=1)
```

```python
# Renombrar la columna unnamed: 1 a Region/Country
df = df.rename(columns={"Unnamed: 1": "Region/Country"})
df.head()
```

```python
# Eliminar columnas no necesarias
df = df.drop(columns=['System of trade', 
                      'System of trade footnote', 
                      'Footnotes', 
                      'Source'])
```

```python
df.head(9)
```

```python
# Filtrar registros con Year > 2015
df_filtrado = df[df["Year"] > 2015]
```

```python
# Lista de valores que conservar
keep_rows = [
    'Total, all countries or areas',
    'Africa',
    'Northern Africa',
    'Sub-Saharan Africa',
    'Eastern Africa',
    'Middle Africa',
    'Southern Africa',
    'Western Africa',
    'Americas',
    'Northern America',
    'Latin America & the Caribbean',
    'Caribbean',
    'Central America',
    'South America',
    'Asia',
    'Central Asia',
    'Eastern Asia',
    'South-Eastern Asia',
    'Southern Asia',
    'Western Asia',
    'Europe',
    'Eastern Europe',
    'Northern Europe',
    'Southern Europe',
    'Western Europe',
    'Oceania',
    'Mexico',
    'United States',
    'Canada',
    'Brazil',
    'Colombia',
    'Argentina',
    'Chile'
]

# Filtrar el DataFrame
df_filtrado = df_filtrado[df_filtrado['Region/Country'].isin(keep_rows)]
```

```python
df_filtrado.shape
```

```python
df_filtrado.head()
```

```python
df_filtrado.tail()
```

```python
import pandas as pd
import plotly.express as px

# Convertir 'Value' a numérico (quitando comas)
df_filtrado['Value'] = df_filtrado['Value'].astype(str).str.replace(',', '').astype(float)

# Filtrar solo Imports CIF
imports = df_filtrado[df_filtrado['Series'] == 'Imports CIF (millions of US dollars)']

# Seleccionar México y Argentina/Brasil/Chile/Colombia
subset = imports[imports['Region/Country'].isin(['Mexico', 'Argentina', 'Brazil', 'Chile', 'Colombia'])]

# Gráfico de barras
fig = px.bar(
    subset,
    x="Year",
    y="Value",
    color="Region/Country",
    barmode="group",
    title="Importaciones CIF (millones de dólares USD): Mexico vs Argentina/Brasil/Chile/Colombia",
    labels={"Value": "Importaciones CIF (millones de dólares USD)", "Year": "Año"}
)
fig.update_traces(texttemplate='%{y:.3s}', textposition='outside')
fig.show()
```

```python
# Filtrar solo Exports FOB
exports = df_filtrado[df_filtrado['Series'] == 'Exports FOB (millions of US dollars)']

# Seleccionar México y Argentina/Brasil/Chile/Colombia
subset = exports[exports['Region/Country'].isin(['Mexico', 'Argentina', 'Brazil', 'Chile', 'Colombia'])]

# Gráfico de barras
fig = px.bar(
    subset,
    x="Year",
    y="Value",
    color="Region/Country",
    barmode="group",
    title="Exportaciones FOB (millones de dóalres USD): Mexico vs Argentina/Brasil/Chile/Colombia",
    labels={"Value": "Exportaciones FOB (millones de dóalres USD)", "Year": "Año"}
)
fig.update_traces(texttemplate='%{y:.3s}', textposition='outside')
fig.show()
```

```python
# Filtrar solo Balance importaciones/exportaciones
balance = df_filtrado[df_filtrado['Series'] == 'Balance imports/exports (millions of US dollars)']

# Seleccionar México y Argentina/Brasil/Chile/Colombia
subset = balance[balance['Region/Country'].isin(['Mexico', 'Argentina', 'Brazil', 'Chile', 'Colombia'])]

# Gráfico de barras
fig = px.bar(
    subset,
    x="Year",
    y="Value",
    color="Region/Country",
    barmode="group",
    title="Balance Comercial (millones de dólares USD): Mexico vs Argentina/Brasil/Chile/Colombia",
    labels={"Value": "Balance Comercial (millones de dólares USD)", "Year": "Año"}
)
fig.update_traces(texttemplate='%{y:.3s}', textposition='outside')
fig.show()
```
